```python
#REDACTION PROTOCOL
ENGINEER = "SINUES"
byte_val = 83
loops = 11
seed = 8 
redacted_output = ""

for char in ENGINEER.upper():
    if char in "AEIOU":
        redacted_output += str(seed)
    else:
        redacted_output += char

print(f"Original Surname: {ENGINEER}")
print(f"SEED_NUM (B): {seed}")
print(f"Redacted Output: {redacted_output}")
```

```python
A = byte_val 
B = seed 
C = loops    

verification_hash = (A * B) - C
print(f"First Letter ASCII (A): {A}")
print(f"SEED_NUM (B): {B}")
print(f"Total Loops (C): {C}")
print(f"Forensic Signature: {verification_hash}")
```
